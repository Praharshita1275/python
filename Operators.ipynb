```python
#On performing floor division a//b, result is the largest integer which is
#less than or equal to the quotient.// is called floor division operator.
print(10//3)  #yeilds 3
print(-10//3)  #yeilds -4
print(10//-3)  #yeilds -4
print(-10//-3)  #yeilds 3
print(3//10)  #yeilds 0
print(3//-10)  #yeilds -1
print(-3//10)  #yeilds -1
print(-3//-10)  #yeilds 0
```

```python
n = int(input("Enter a number"))
sum=0
while(n>0):
  digit = n%10
  sum = sum+digit
  n = n//10
print(sum)
```

```python
#Armstrong
n = int(input("Enter a number "))
sum = 0
temp = n
while (n>0):
  digit = n%10
  sum = sum+digit*digit*digit
  n = n//10
print(sum)
if temp ==sum:
  print("The no is Armstrong")
else:
  print("Not armstrong")
```

```python
#Finding reverse of a number
n = int(input("Enter a number "))
temp = n
reverse = 0
while(n>0):
  d = n%10
  reverse = reverse*10+d
  n = n//10
  print(reverse)
if temp ==reverse:
  print("The number is a palindrome")
else:
  print("NOt a palindrome")
```
