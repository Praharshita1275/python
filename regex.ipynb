```python
import re

def is_valid_email(email):
    pattern = r'^[a-zA-Z0-9._%+-]+@[a-zA-Z0-9.-]+\.[a-zA-Z]+$'

    if re.match(pattern, email):
        return True
    else:
        return False

print(is_valid_email("john.doe@example.com"))
print(is_valid_email("invalid_email@.com"))
print(is_valid_email("user-name123@example-domain.com"))
```

```python
#2.Extract Phone Numbers:
#Write a function that takes a string as input and returns a list of all valid phone numbers found in the string. A valid phone number should consist of 10 digits (with optional dashes or spaces between them).
#Example:
#Input: "Contact us at 123-456-7890 or 555 123 4567"
#Output: ['123-456-7890', '555 123 4567']

 import re

def extract_phone_numbers(input_string):
    pattern = r'\b(?:\d{3}[-\s]?\d{3}[-\s]?\d{4})\b'

    phone_numbers = re.findall(pattern, input_string)

    return phone_numbers
input_string = "Contact us at 123-456-7890 or 555 123 4567"
output = extract_phone_numbers(input_string)
print(output)
```

```python
import re
def is_valid_password(password):
    if len(password) < 8:
        return False
    if not re.search(r'[A-Z]', password) or not re.search(r'[a-z]', password) or not re.search(r'\d', password):
        return False
    return True

password = "P@ssw0rd"
print(is_valid_password(password))
```

```python
import re
def find_long_words(text):
    pattern = re.compile(r'\b\w{4,}\b')
    long_words = pattern.findall(text)
    return long_words
text = "This is a sample text with some long words like Python, programming, and algorithm."
result = find_long_words(text)
print(result)
```
